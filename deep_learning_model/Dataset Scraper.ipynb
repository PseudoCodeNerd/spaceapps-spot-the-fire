```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd 
```

```python
data = {'values': [], 'dates': []}
```

```python
for year in range(2000,2021):
    page = requests.get("https://neo.sci.gsfc.nasa.gov/view.php?datasetId=MODAL2_E_SKY_WV&year="+str(year))
    soup = BeautifulSoup(page.content, 'html.parser')
    for x in soup.find_all('a'):
        try:
            if re.match('viewDataset',x['onclick']): 
                value = x['onclick'][13:20]
                date = x['onclick'][23:33]
                data['values'].append(value)
                data['dates'].append(date)
        except:
            pass
print(data)
```

```python
print(data)
```

```python
for index in range(len(data['values'])):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:20.0) Gecko/20100101 Firefox/20.0'}
    url = "https://neo.sci.gsfc.nasa.gov/servlet/RenderData?si="+data['values'][index]+"&cs=rgb&format=JPEG&width=720&height=360"
    r = requests.get(url, allow_redirects=True, headers=headers)
    open("Water Vapour/"+ data['dates'][index] + ".jpg", 'wb').write(r.content)
```

```python
#Initialize DataFrame
df=pd.DataFrame(columns=['location','lat','lon'])
df['location']=['LA','NY','LV']
df['lat']=[34.05,40.71,36.16]
df['lon']=[-118.24,-74.00,-115.14]

#New point Reno 39.53,-119.81
newlat=39.53
newlon=-119.81

#Import trig stuff from math
from math import sin, cos, sqrt, atan2,radians

#Distance function between two lat/lon
def getDist(lat1,lon1,lat2,lon2):
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    return R * c

#Apply distance function to dataframe
df['dist']=list(map(lambda k: getDist(df.loc[k]['lat'],df.loc[k]['lon'],newlat,newlon), df.index))

#This will give all locations within radius of 600 km
df[df['dist']<500]
```
